$$
\vec{A} = re^{-r^2} \vec{e} _ r
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 
r = np.linspace(0.01, 2.5, 10)  # 
theta = np.linspace(0, 2*np.pi, 30)  # 
R, Theta = np.meshgrid(r, theta)    # rtheta

# 
Ur = r * np.exp(-r**2)  # r
Utheta = 0  # theta

# 
X = R * np.cos(Theta)
Y = R * np.sin(Theta)
U = Ur * np.cos(Theta) - Utheta * np.sin(Theta)
V = Ur * np.sin(Theta) + Utheta * np.cos(Theta)

# 
plt.figure(figsize=(10, 10))
plt.quiver(X, Y, U, V)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('')
plt.grid()
plt.axis('equal')
plt.show()
```
